```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def difference(a):
    c = len(a)-1
    b = np.ones(c)
    b[:] = a[:-1] - a[1:]
    return b
```

```python
t = np.genfromtxt("triangular_perc_3.txt")
t_t = np.genfromtxt("triangular_perc_testing_3.txt")

k = 39
n = 4

plt.plot(t[k*n:k*(n+1), 0], t[k*n:k*(n+1), 1], "ro")
plt.plot(t_t[:, 0], t_t[:, 1], "bx")
```

```python
b = difference(t[k*n:k*(n+1), 1])
plt.plot(np.arange(0, len(b)), b)
```

```python
t = np.genfromtxt("triangular_perc_2.txt")

k = 39
n = 2

plt.plot(t[k*n:k*(n+1), 0], t[k*n:k*(n+1), 1], "ro")
```

```python
res = np.ones((k-1, (len(t)-1)//k))

print(np.shape(res))

for i in range((len(t)-1)//k):
    res[:, i] = np.abs(difference(t[k*i:k*(i+1), 1]))


plt.imshow(res, aspect="auto")
plt.show()
```
