```python
import string
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as opt
import time
import random as rn
import matplotlib.colors as colors
import matplotlib.cbook as cbook
from matplotlib import cm
import matplotlib.patches as mpl_patches
from matplotlib import rcParams
from io import StringIO
import matplotlib.tri as mtri
colormap = cm.inferno
plt.rcParams.update({'font.size': 20})



def squ_conv(n, L=100):
    
        x = []
        y = []
        for i in n:
            x.append(i % L)
            y.append(int(i/L))
        
        return x, y

f = open("snapshot_L100_0.100_L100_phi1.05_3.txt")
sq3 = []
for line in f:
    sq3.append(np.loadtxt(StringIO(line), dtype=int))

fig = plt.figure()
plt.xticks([])
plt.yticks([])
plt.scatter(x=squ_conv(sq3[0][::3])[0], y=squ_conv(sq3[0][::3])[1], c=sq3[0][1::3], cmap="Greens", s=4, vmin=0, vmax=3, marker="s")
plt.axis([0, 40, 0, 40])
```

```python
fig = plt.figure()
plt.xticks([])
plt.yticks([])
plt.scatter(x=squ_conv(sq3[0][::3])[0], y=squ_conv(sq3[0][::3])[1], c=sq3[0][2::3]+1, cmap="Reds", s=4, vmin=0, vmax=2, marker="s")
plt.axis([0, 40, 0, 40])
```
